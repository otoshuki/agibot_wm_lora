```python
import os
import glob
import h5py
import json
import numpy as np
from tqdm import tqdm

# Point this at your train split root(s)
DATA_ROOTS = ["/workspace/AgiBotWM/WorldModel/mnt/public/chenshengcong/dataset/iros_challenge_2025_acwm/train"]

# Collect every proprio_stats.h5 under the train split
h5_files = []
for root in DATA_ROOTS:
    h5_files.extend(glob.glob(os.path.join(root, "*", "proprio_stats.h5")))
print(f"Found {len(h5_files)} episodes")

# Streaming mean/variance via Welford's algorithm — avoids loading everything into RAM
count = 0
mean  = np.zeros(12, dtype=np.float64)
M2    = np.zeros(12, dtype=np.float64)

mins = np.full(12,  np.inf, dtype=np.float64)
maxs = np.full(12, -np.inf, dtype=np.float64)

for f in tqdm(h5_files):
    try:
        with h5py.File(f, "r") as fid:
            w = np.asarray(fid["state/end/wrench"], dtype=np.float64)  # (T, 12)
            print(w)
    except Exception as e:
        print(f"skip {f}: {e}")
        continue

    if w.ndim != 2 or w.shape[1] != 12:
        print(f"skip {f}: unexpected shape {w.shape}")
        continue

    mins = np.minimum(mins, w.min(axis=0))
    maxs = np.maximum(maxs, w.max(axis=0))

    # Welford update per frame
    for row in w:
        count += 1
        delta  = row - mean
        mean  += delta / count
        delta2 = row - mean
        M2    += delta * delta2

var = M2 / max(count - 1, 1)
std = np.sqrt(var)

stats = {
    "wrench_mean": mean.tolist(),
    "wrench_std":  std.tolist(),
    "wrench_min":  mins.tolist(),
    "wrench_max":  maxs.tolist(),
    "wrench_count": int(count),
}
print(json.dumps(stats, indent=2))

with open("wrench_stats.json", "w") as f:
    json.dump(stats, f, indent=2)
```
